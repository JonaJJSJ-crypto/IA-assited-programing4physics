```python
# prompt: Grfica el potencial eléctrico de dos cargas puntual

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Constants
k = 8.98755e9 # Coulomb's constant

# Charges and positions
q1 = 2.5  # Charge 1 (Coulombs)
x1, y1 = -0.1, 0  # Position of charge 1
q2 = -2.5  # Charge 2 (Coulombs)
x2, y2 = 0.1, 0   # Position of charge 2

# Create a grid
x = np.linspace(-0.15, 0.15, 100)
y = np.linspace(-0.1, 0.1, 100)
X, Y = np.meshgrid(x, y)

# Calculate the potential at each point
V = k * (q1 / np.sqrt((X - x1)**2 + (Y - y1)**2) +
         q2 / np.sqrt((X - x2)**2 + (Y - y2)**2))


# Plot the equipotential lines
plt.figure(figsize=(8, 6))
contours = plt.contour(X, Y, V, 200)  # Adjust the number of contours as needed
#plt.clabel(contours, inline=1, fontsize=8)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Equipotential Lines for Two Point Charges')

# Plot the charges
plt.plot(x1, y1, 'ro', label=f'q1 = {q1:.2e} C')
plt.plot(x2, y2, 'bo', label=f'q2 = {q2:.2e} C')
plt.legend()
plt.grid(True)
plt.show()


# Calculate the potential at each point
V = k * (q1 / np.sqrt((X - x1)**2 + (Y - y1)**2) +
         q2 / np.sqrt((X - x2)**2 + (Y - y2)**2))

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the potential surface
ax.plot_surface(X, Y, V, cmap='viridis')

# Set labels and title
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
ax.set_zlabel('Electric Potential (V)')
ax.set_title('Electric Potential of Two Point Charges')

# Show the plot
plt.show()
```
