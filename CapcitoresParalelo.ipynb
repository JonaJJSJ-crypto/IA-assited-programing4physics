```python
# prompt: Genera una tabla en pandas que me permita registrar voltajes y tiempo

import pandas as pd

# Crear un diccionario para almacenar los datos
data = {'Tiempo (s)': [0,2.66,5.54,9.60,12.72], 'Voltaje (V)': [10,8,6,4,3]}

# Crear un DataFrame vacío
df = pd.DataFrame(data)

# Puedes agregar datos de la siguiente manera:
# df.loc[len(df)] = [tiempo, voltaje]

# Imprimir la tabla
df
```

```python
# prompt: Grafica el voltaje en funcion del tiempo

import matplotlib.pyplot as plt

# Graficar el voltaje en función del tiempo
plt.plot(df['Tiempo (s)'], df['Voltaje (V)'], marker='o', linestyle='-')
plt.xlabel('Tiempo (s)')
plt.ylabel('Voltaje (V)')
plt.title('Voltaje vs. Tiempo')
plt.grid(True)
plt.show()
```

```python
# prompt: Genera un ajuste lineal con una curva exponencial decreciente y calcula la constante de tiempo

import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Define the exponential decay function
def exponential_decay(x, a, b, c):
    return a * np.exp(-b * x) + c

# Extract data from the DataFrame
tiempo = df['Tiempo (s)'].values
voltaje = df['Voltaje (V)'].values

# Perform the curve fitting
# Provide initial guess for the parameters (a, b, c)
initial_guess = [10, 0.5, 0]
params, covariance = curve_fit(exponential_decay, tiempo, voltaje, p0=initial_guess)

# Extract the optimized parameters
a_opt, b_opt, c_opt = params

# Calculate the time constant (tau)
# For an exponential decay of the form A * exp(-b * t) + C, the time constant is 1/b
tau = 1 / b_opt

# Print the optimized parameters and the time constant
print(f"Optimized parameters:")
print(f"  a = {a_opt:.4f}")
print(f"  b = {b_opt:.4f}")
print(f"  c = {c_opt:.4f}")
print(f"\nTime constant (tau) = {tau:.4f} s")

# Generate points for the fitted curve
tiempo_fit = np.linspace(tiempo.min(), tiempo.max(), 100)
voltaje_fit = exponential_decay(tiempo_fit, a_opt, b_opt, c_opt)

# Plot the original data and the fitted curve
plt.figure()
plt.plot(tiempo, voltaje, marker='o', linestyle='-', label='Datos originales')
plt.plot(tiempo_fit, voltaje_fit, linestyle='--', color='red', label='Ajuste exponencial')
plt.xlabel('Tiempo (s)')
plt.ylabel('Voltaje (V)')
plt.title('Ajuste exponencial a datos de Voltaje vs. Tiempo')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# prompt: Calcula la capacitancia del Circuito en funcion de la constante de tiempo y una ressitencia de 100000 ohmios

# Resistencia dada en Ohmios
R = 100000  # 100 kOhmios

# Calcular la capacitancia (C) usando la fórmula tau = R * C
# Despejando C: C = tau / R
C = tau / R

# Imprimir el valor de la capacitancia
print(f"\nLa capacitancia (C) es = {C:.8f} Faradios")
print(f"La capacitancia (C) es = {C * 1e6:.4f} microFaradios")
```
