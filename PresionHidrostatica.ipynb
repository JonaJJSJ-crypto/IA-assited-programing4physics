```python
import pandas as pd

data = {"dh(m)":[0.014,0.036,0.054,0.082,0.10],
        "y(m)":[0.02,0.04,0.06,0.08,0.1]}

df = pd.DataFrame(data)

df["Pm(Pa)"] = 1000 * 9.8 * df["dh(m)"] #Presion manometro
df["Ph(Pa)"] = 1000 * 9.8 * df["y(m)"] #Presion Hidrostatica

df
```

```python
# prompt: Genera un template para dibujar en matplotlib

import matplotlib.pyplot as plt

plt.figure(figsize=(4, 4))  # Ajusta el tamaño de la figura

plt.plot(df["Pm(Pa)"], df["Ph(Pa)"], marker='o', linestyle='-', color="red")  # Dibuja la gráfica

plt.title("Dispersion de presiones")
plt.xlabel("P-manometrica (Pa)")
plt.ylabel("P-hidrostatica(Pa)")
plt.grid(True)  # Agrega una cuadrícula
plt.show()
```
