```python
# prompt: Gera un modelo de tabla en pandas con dicionario

import pandas as pd

data = {'Masa(g)': [50,100,150,200,250,300,350,350,300,250,200,150,100,50],
        'y(m)': [0.452,0.515,.579,.641,.702,.765,.826,.826,.765,.702,.640,.578,.515,.453]}
df = pd.DataFrame(data)
df['Masa(kg)'] = df['Masa(g)']/1000

df
```

```python
# prompt: grafica la mas aen funcion de y
import matplotlib.pyplot as plt

# Create the plot
plt.scatter(df['y(m)'], df['Masa(kg)'], marker='d', color="blue")

# Add labels and title
plt.xlabel('y [m]')
plt.ylabel('Fuerza/(9.8) [N]')
plt.title('Fuerza en función de y')
plt.grid(True)

# Display the plot
plt.show()
```

```python
# prompt: Usa el metodo de minimos cuadrados para calcular la pendiente de la funcion presentada en el grafico y traza una linea de tendencia

import matplotlib.pyplot as plt
import numpy as np

# Calculate the slope and intercept using least squares
x = df['y(m)']
y = df['Masa(kg)']*9.8
slope, intercept = np.polyfit(x, y, 1)

# Create the trendline
trendline = slope * x + intercept

# Plot the data and trendline
plt.scatter(x, y, marker='d', color="blue", label='Datos')
plt.plot(x, trendline, color='red', label=f'Trendline: y = {slope:.2f}x + {intercept:.2f}')

# Add labels, title, and legend
plt.xlabel('y [m]')
plt.ylabel('Fuerza/(9.8) [N]')
plt.title('Fuerza en función de y')
plt.grid(True)
plt.legend()


# Display the plot
plt.show()

print(f"La pendiente de la línea de tendencia es: {slope:.2f}")
```
