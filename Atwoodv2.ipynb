```python
# prompt: ahora vas a generar otra columna llamada "F(N)" que es igual a dM(g)*9,81 y lo redondeas a 3 decimales

import pandas as pd

data = {'dM(g)': [0,4, 8, 12,16,20], 'MT(g)': [490,490, 490, 490,490,490],'tp(s)': [0,5.33,2.79,2.19,1.74,1.50]}
df = pd.DataFrame(data)

df['F(N)'] = df['dM(g)'] * 9.81
df['F(N)'] = df['F(N)'].round(3)

df
```

```python
# prompt: ahora vas a generar otra columna llamada "a(m/s^2)" y es igual a (dM/MT)*9,81

import pandas as pd

data = {'dM(g)': [4, 8, 12,16,20], 'MT(g)': [490, 490, 490,490,490],'tp(s)': [5.33,2.79,2.19,1.74,1.50]}
df = pd.DataFrame(data)

df['F(N)'] = df['dM(g)'] * 9.81
df['F(N)'] = df['F(N)'].round(3)

df['a_teo(m/s^2)'] = (df['dM(g)'] / df['MT(g)']) * 9.81
df['a_teo(m/s^2)'] = df['a_teo(m/s^2)'].round(3)

df['a_exp(m/s^2)'] = (1 / df['tp(s)']**2)
df['a_exp(m/s^2)'] = df['a_exp(m/s^2)'].round(3)

df
```

```python
# prompt: grafica la aceleracion experimental en funcion de la Fuerza y la aceleracion teorica en funcion de la fuerza

import pandas as pd
import matplotlib.pyplot as plt

# Data provided in the original code
data = {'dM(g)': [4, 8, 12,16,20], 'MT(g)': [490, 490, 490,490,490],'tp(s)': [5.33,2.79,2.19,1.74,1.50]}
df = pd.DataFrame(data)

df['F(N)'] = df['dM(g)'] * 9.81
df['F(N)'] = df['F(N)'].round(3)

df['a_teo(m/s^2)'] = (df['dM(g)'] / df['MT(g)']) * 9.81
df['a_teo(m/s^2)'] = df['a_teo(m/s^2)'].round(3)

df['a_exp(m/s^2)'] = (1 / df['tp(s)']**2)
df['a_exp(m/s^2)'] = df['a_exp(m/s^2)'].round(3)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot experimental acceleration vs. force
plt.plot(df['F(N)'], df['a_exp(m/s^2)'], marker='o', linestyle='-', color='blue', label='Experimental Acceleration')

# Plot theoretical acceleration vs. force
plt.plot(df['F(N)'], df['a_teo(m/s^2)'], marker='x', linestyle='--', color='red', label='Theoretical Acceleration')

# Customize the plot
plt.xlabel('Force (N)', fontsize=12)
plt.ylabel('Acceleration (m/s^2)', fontsize=12)
plt.title('Experimental vs. Theoretical Acceleration vs. Force', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```
