```python
# prompt: Gráfica la posición en función del tiempo

import pandas as pd
import matplotlib.pyplot as plt

data = {'posicion(cm)': [40, 80, 120, 160, 200],
        'tiempo(s)': [1.76, 2.656, 3.39, 3.816, 4.46]}

df = pd.DataFrame(data)

plt.plot(df['tiempo(s)'], df['posicion(cm)'])
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (cm)')
plt.title('x(t) Mrua')
plt.grid(True)
plt.show()
```

```python
# prompt: Gráfica la posición en función del tiempo elevado al cuadrado

import pandas as pd
import matplotlib.pyplot as plt

data = {'posicion(cm)': [40, 80, 120, 160, 200],
        'tiempo(s)': [1.76, 2.656, 3.39, 3.816, 4.46]}

df = pd.DataFrame(data)

# Calculate the square of the time
df['tiempo(s)^2'] = df['tiempo(s)']**2

# Create the plot
plt.plot(df['tiempo(s)^2'], df['posicion(cm)'], marker='o') # Added markers for better visualization
plt.xlabel('Tiempo² (s²)')
plt.ylabel('Posición (cm)')
plt.title('Posición vs. Tiempo²')
plt.grid(True)
plt.show()
```

```python
# prompt: Realiza una regresion lineal de la posición en función del tiempo al cuadrado. Y muestra la línea de tendencies en una gráfica. También imprime la ecuacion obtenida

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {'posicion(cm)': [40, 80, 120, 160, 200],
        'tiempo(s)': [1.76, 2.656, 3.39, 3.816, 4.46]}

df = pd.DataFrame(data)

# Calculate the square of the time
df['tiempo(s)^2'] = df['tiempo(s)']**2

# Perform linear regression
x = df['tiempo(s)^2']
y = df['posicion(cm)']
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)

# Print the equation
print(f"Ecuación de la recta de tendencia: y = {coefficients[0]:.2f}x + {coefficients[1]:.2f}")

# Generate points for the trend line
x_trend = np.linspace(min(x), max(x), 100)  # Increased number of points
y_trend = polynomial(x_trend)


# Create the plot
plt.plot(df['tiempo(s)^2'], df['posicion(cm)'], marker='o', linestyle='none', label='Datos') # Added markers for better visualization and removed line
plt.plot(x_trend, y_trend, color='red', label='Regresión lineal') # Added line for trend line
plt.xlabel('Tiempo² (s²)')
plt.ylabel('Posición (cm)')
plt.title('Posición vs. Tiempo²')
plt.grid(True)
plt.legend() # Display legend
plt.show()
```
