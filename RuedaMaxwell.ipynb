```python
# prompt: Generame un modelo de tabla de pandas

import pandas as pd

# Crea un diccionario con los datos de la tabla
data = {
    "h(m)":[0.100,0.200,0.300,0.400,0.500],
    'tp(s)': [0.084,0.064,0.049,0.042,0.035]
}

# Crea un DataFrame de pandas a partir del diccionario
df = pd.DataFrame(data)

# Muestra la tabla
df
```

```python
df["Ep(J)"] = df["h(m)"] * 9.8 * 0.436

df["v(m/s)"] = 0.05 / df["tp(s)"]

df["Ec(J)"] = df["v(m/s)"]**2 * 0.436 * 0.5

df["Er(J)"] = 0.5 * 9.84e-4 * 4 / df["tp(s)"]**2

df["T(J)"] = df["Ec(J)"] + df["Er(J)"]

df
```

```python
df["%"] = (df["T(J)"]-df["Ep(J)"]) / df["Ep(J)"]  * 100
df
```
