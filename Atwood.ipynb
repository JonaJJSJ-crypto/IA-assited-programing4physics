```python
# prompt: Genera una tabla de pandas con aceleración y Fuerza

import pandas as pd
import numpy as np

# Crear datos de ejemplo
aceleracion = [1,2,3,4,5] # Aceleración en m/s^2
fuerza = [3,5,7,9,10] # Fuerza en Newtons

# Crear el DataFrame de pandas
df = pd.DataFrame({
    'Aceleracion (m/s^2)': aceleracion,
    'Fuerza (N)': fuerza
})

# Imprimir la tabla
df
```

```python
# prompt: Grafica la fuerza en función de la aceleración

import matplotlib.pyplot as plt

# Graficar la fuerza en función de la aceleración
plt.figure(figsize=(8, 6))
plt.plot(df['Aceleracion (m/s^2)'], df['Fuerza (N)'], marker='o', linestyle='-')
plt.title('Fuerza vs Aceleración')
plt.xlabel('Aceleración (m/s^2)')
plt.ylabel('Fuerza (N)')
plt.grid(True)
plt.show()
```

```python
# prompt: Calcula la pendiente del gráfico y muestra línea de tendencia

import matplotlib.pyplot as plt
import numpy as np
# Calcula la pendiente utilizando numpy.polyfit
# Se ajusta un polinomio de grado 1 (línea) a los datos
slope, intercept = np.polyfit(df['Aceleracion (m/s^2)'], df['Fuerza (N)'], 1)

# Imprime la pendiente
print(f"La pendiente del gráfico (la masa) es: {slope:.2f}")

# Genera puntos para la línea de tendencia
x = df['Aceleracion (m/s^2)']
y = slope * x + intercept

# Grafica los datos originales y la línea de tendencia
plt.figure(figsize=(8, 6))
plt.plot(df['Aceleracion (m/s^2)'], df['Fuerza (N)'], marker='o', linestyle='-', label='Datos originales')
plt.plot(x, y, color='red', linestyle='--', label=f'Línea de tendencia (y = {slope:.2f}x + {intercept:.2f})')
plt.title('Fuerza vs Aceleración con Línea de Tendencia')
plt.xlabel('Aceleración (m/s^2)')
plt.ylabel('Fuerza (N)')
plt.grid(True)
plt.legend()
plt.show()
```
