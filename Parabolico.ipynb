```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función para calcular el campo magnético de un dipolo
def campo_magnetico_dipolo(x, y, z, mu0, m):
  """
  Calcula el campo magnético de un dipolo magnético.

  Args:
    x, y, z: Coordenadas del punto donde se calcula el campo.
    mu0: Permeabilidad magnética del vacío.
    m: Momento dipolar magnético.

  Returns:
    Una tupla con los componentes Bx, By, Bz del campo magnético.
  """
  # Calculate distance to the point
  r = np.sqrt(x**2 + y**2 + z**2)

  # Calculate dot product using broadcasting
  dot_product = m[0] * x + m[1] * y + m[2] * z

  # Calculate magnetic field components using broadcasting
  Bx = (mu0 / (4 * np.pi)) * (3 * x * dot_product / r**5 - m[0] / r**3)
  By = (mu0 / (4 * np.pi)) * (3 * y * dot_product / r**5 - m[1] / r**3)
  Bz = (mu0 / (4 * np.pi)) * (3 * z * dot_product / r**5 - m[2] / r**3)

  return Bx, By, Bz


# Parámetros del imán
mu0 = 4 * np.pi * 1e-7  # Permeabilidad magnética del vacío
m = np.array([0, 1, 0])  # Momento dipolar magnético (ejemplo: imán apuntando hacia arriba)

# Crear una malla de puntos para representar el espacio
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)  # Plano xy

# Calcular el campo magnético en cada punto de la malla
Bx, By, Bz = campo_magnetico_dipolo(X, Y, Z, mu0, m)


Treshold = 1e-5
for i in range(len(By)):
  for j in range(len(By[i])):
    if np.abs(By[i][j]) > Treshold:
      By[i][j] = 0
      Bx[i][j] = 0
      Bz[i][j] = 0
for i in range(len(Bx)):
  for j in range(len(Bx[i])):
    if np.abs(Bx[i][j]) > Treshold:
      By[i][j] = 0
      Bx[i][j] = 0
      Bz[i][j] = 0
for i in range(len(Bz)):
  for j in range(len(Bz[i])):
    if np.abs(Bz[i][j]) > Treshold:
      By[i][j] = 0
      Bx[i][j] = 0
      Bz[i][j] = 0
# Graficar el campo magnético usando quiver
plt.figure(figsize=(8, 8))
plt.quiver(X, Y, Bx, By, color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Campo magnético de un imán (plano xy)')
plt.grid(True)
plt.show()
```
